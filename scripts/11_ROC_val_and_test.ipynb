```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 5))
lw = 2
def plot_roc_curve(fpr, tpr, label=None, color=None):
    
    font1 = {'family' : 'Arial',
    'weight' : 'bold',
    'size' : 15,
    }
    font2 = {'family' : 'Arial',
    'weight' : 'normal',
    'size' : 13,
    }
    
    plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate (FPR)', font2)
    plt.ylabel('True Positive Rate (TPR)', font2)
    plt.title('ROC Curves for Validation Set', font1)

    step_x = list(fpr)
    step_x.append(1.0)  # Add endpoint for the last step
    step_y = list(tpr)
    step_y.append(tpr[-1])  # Add endpoint for the last step
    
    plt.step(step_x, step_y, linewidth=2, label=label, color = color, alpha=0.7)
    plt.legend(loc="lower right")

plot_roc_curve(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC = {roc_auc_lr:.3f})', color='b')
plot_roc_curve(fpr_nb, tpr_nb, label=f'Naive Bayes (AUC = {roc_auc_nb:.3f})', color='g')
plot_roc_curve(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {roc_auc_rf:.3f})', color='k')
plot_roc_curve(fpr_svm, tpr_svm, label=f'Support Vector Machine (AUC = {roc_auc_svm:.3f})', color='c')
plot_roc_curve(fpr_xgb, tpr_xgb, label=f'eXtreme Gradient Boost (AUC = {roc_auc_xgb:.3f})', color='m')
plot_roc_curve(fpr_cnn, tpr_cnn, label=f'Convoluntional Neural Network (AUC = {roc_auc_cnn:.3f})', color='y')
plot_roc_curve(fpr_lstm, tpr_lstm, label=f'Long-Short Term Memory (AUC = {roc_auc_lstm:.3f})', color='r')
plot_roc_curve(fpr_resnet, tpr_resnet, label=f'Residual Neural Network (AUC = {roc_auc_resnet:.3f})', color='mediumpurple')
plot_roc_curve(fpr_transformer, tpr_transformer, label=f'1d-Transformer (AUC = {roc_auc_transformer:.3f})', color='brown')

plt.savefig('/Users/jiaming/Desktop/Lab2/datas/ROC/ROC_val.pdf', format='pdf')
plt.show()
```
