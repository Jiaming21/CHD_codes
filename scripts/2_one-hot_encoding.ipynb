```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/jiaming/Desktop/Lab2/datas/framingham_without_nas.csv')
```

```python
data
```

```python
cat = data.loc[:, ['male', 'education','currentSmoker', 'BPMeds', 'prevalentStroke','prevalentHyp', 'diabetes', 'TenYearCHD']]
cat.head()
```

```python
# convert them to one hot encoding. 
cat_str = cat.astype(str) # nominal to string
cat_oh = pd.get_dummies(cat_str)
cat_oh = cat_oh.drop('TenYearCHD_0', axis = 1) # drop unnecessry column 
cat_oh.head()
```

```python
data_new = data.drop(['male', 'education','currentSmoker', 'BPMeds', 'prevalentStroke','prevalentHyp', 'diabetes', 'TenYearCHD'], axis =1)
data_new.head()
```

```python
data_cb = pd.concat([data_new, cat_oh], axis = 1) #Combine categorical features to dataset
data_cb.head()
```

```python
# Lets rename target feature
data_cb.rename({'TenYearCHD_1':'target'}, axis=1, inplace=True)
data_cb.head()
```
